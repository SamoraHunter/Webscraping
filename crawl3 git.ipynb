```python
import re
import requests
from urllib.parse import urlsplit
from collections import deque
from bs4 import BeautifulSoup
import pandas as pd
import pdb
#from google.colab import files
#import pdb

urlList = open("10AIList.txt").readlines()

status = 'Unknown'

emails = set()
%pdb
from IPython.core.debugger import set_trace



for url in urlList: 
    
    pdb.set_trace()
    original_url = url
    
    #original_url = 'https://www.advicerobo.com/'
    

    unscraped = deque([original_url])  

    scraped = set()  


    while len(unscraped):
        url = unscraped.popleft()  
        scraped.add(url)

        parts = urlsplit(url)

        base_url = "{0.scheme}://{0.netloc}".format(parts)
        if '/' in parts.path:
          path = url[:url.rfind('/')+1]
        else:
          path = url

        print("Crawling URL %s" % url, status)
        try:
            response = requests.get(url)
        except (requests.exceptions.MissingSchema, requests.exceptions.ConnectionError):
            continue

        new_emails = set(re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.com", response.text, re.I))
        
        
        Preupdate = emails
        emails.update(new_emails) 
        Postupdate = emails
        
        if (Preupdate == Postupdate):
            #print("Not found")
            status = 'Missing'
        else:
            #print("Found")
            status = 'Found!!'
            

        soup = BeautifulSoup(response.text, 'lxml')

    for anchor in soup.find_all("a"):
        if "href" in anchor.attrs:
            link = anchor.attrs["href"]
        else:
            link = ''

            if link.startswith('/'):
                link = base_url + link

            elif not link.startswith('http'):
                link = path + link

        if not link.endswith(".gz"):
            if not link in unscraped and not link in scraped:
                  unscraped.append(link)  

emailstring =  sorted(emails)

df = pd.DataFrame(emailstring, columns=["Email"]) 

#df.append(emailstring)
    

   

 #df.to_csv('crawl2.csv')
#files.download("email.csv")
```
